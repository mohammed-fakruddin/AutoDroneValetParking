```python
def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image
```

```python
def preProcessImages(inputFolder,preProcessFolder):
    subfolders = [ f.path for f in os.scandir(inputFolder) if f.is_dir() ]
    for directory in subfolders:
        sub_dir = directory.split('\\')[1]
        os.makedirs(os.path.join(preProcessFolder,sub_dir), exist_ok=True)
        _, _, filenames = next(walk(directory))
        for f in filenames:
            dest_file = os.path.join(preProcessFolder,sub_dir, f) 
            source_file = os.path.join(inputFolder,sub_dir,f)
            img = cv2.imread(source_file)
            img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            img_gray = cv2.resize(img, (width,height))
            imshape = img_gray.shape
            vertices = np.array([[(imshape[1]*1/3,0),
                           (1/3*imshape[1],imshape[0]*1/2),(2/3*imshape[1],1/2*imshape[0]),
                          (2/3*imshape[1],0)]],dtype=np.int32)
            img_no_roi = region_of_interest(img_gray,vertices)
            cv2.imwrite(dest_file,img_no_roi)
            
        
```

```python
preProcessImages(input_folder,pre_process_folder)
```
